# Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# lets create sunthetic data points
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=10)
```

```python
# using svc
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear') # here our data is linearly seprable so we can use linear kernel depends on data orientation we can use any kernel
```

```python
svc.fit(X_train,y_train)
```

```python
# prediction
y_pred = svc.predict(X_test)
```

```python
svc.coef_
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
## for multiple overlappd data 
# use different kernels
## rbf kernel
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=10)
svc=SVC(kernel='rbf')
svc.fit(X_train,y_train)
# prediction
y_pred = svc.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred)) # accuracy increases as we different kernals
```

```python
## polynomial
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=10)
svc=SVC(kernel='poly')
svc.fit(X_train,y_train)
# prediction
y_pred = svc.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))  # giving bad accuracy
```

```python
# sigmoid
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=10)
svc=SVC(kernel='sigmoid')
svc.fit(X_train,y_train)
# prediction
y_pred = svc.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred)) # giving bad accuracy
```

## Hyperparameter tunning

- Here we see which parameters are best for our model

```python
## Once you decide a kernel to use we can do hyperparameter tunnig
from sklearn.model_selection import GridSearchCV
# defining parameters range
param_grid = {'C' : [0.1,1,10,100,1000],
              'gamma': [1,0.1,0.01,0.001,0.0001],
              'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(), param_grid=param_grid,cv=5,refit=True,verbose=3 )
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## predictions
y_pred = grid.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
