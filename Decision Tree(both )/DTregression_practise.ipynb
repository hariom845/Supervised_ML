```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
df = load_diabetes()
```

```python
type(df)
```

```python
df.keys()
```

```python
df.feature_names # independent features
```

```python
df.target
```

```python
X = pd.DataFrame(df.data, columns = df.feature_names)
```

```python
y=df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test,y_pred))
```

- r2_score is very less it means overfitting is happend due to max depth

```python
param = {
    'criterion': ['squared_erro','friedman_mse','absolute_error'],
    'splitter' : ['best','random'],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features':['auto', 'sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
model = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(estimator=model, param_grid=param, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(r2_score(y_test, y_pred))

print(mean_squared_error(y_test,y_pred))
```
