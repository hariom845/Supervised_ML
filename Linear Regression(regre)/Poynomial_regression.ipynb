### Polynomial regression
- there is non-linear relation between the features or data points

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline # it is used to show plots in jupitor
```

```python
X= 6*np.random.rand(100,1) - 3 # here 3 is outlier
y = 0.5*X**2+1.5*X +2 +np.random.randn(100,1) # np.random.randn(100,1) is used as outlier

# plt.plot(X,y,'g') ghatiya plot becasue of non linear relation
plt.scatter(X,y,color='r')
plt.xlabel("X dataset")
plt.ylabel('y dataset')
```

```python
# let's first do with simple linear regression
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression_1= LinearRegression()
regression_1.fit(X_train,y_train)
```

```python
print(regression_1.coef_) # having only one parameter or coeficiient
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, regression_1.predict(X_test))
print(score)
```

```python
# lets visualize this model 
plt.plot(X_train,regression_1.predict(X_train), color = 'r')
plt.scatter(X_train, y_train)
plt.xlabel('X dataset')
plt.ylabel('Y or Y_pred')
# as we have more error therefore we increase the degree and moving towards the polynomial regression
```

## Polynomial regression

```python
# lets apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2, include_bias = True)
x_train_poly =  poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
x_train_poly # here we converted into a quadratic equation because we increasing polynomial degree
```

```python
from sklearn.metrics import r2_score
regression = LinearRegression()
regression.fit(x_train_poly, y_train)
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test, y_pred)
print(score)
```

```python
print(regression.coef_) # having three coeficent
```

```python
print(regression.intercept_)
```

```python
plt.scatter(X_train, regression.predict(x_train_poly))
# it is my best fit line for all data points
plt.scatter(X_train, y_train) # it shows all points
```

```python
# lets make degreee 3
poly = PolynomialFeatures(degree = 3, include_bias = True)
x_train_poly =  poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
x_train_poly
```

```python
from sklearn.metrics import r2_score
regression = LinearRegression()
regression.fit(x_train_poly, y_train)
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test, y_pred)
print(score)
```

```python
## prediction for new data
import numpy as np
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_poly = poly.transform(X_new) # transforming data to polynomial data
```

```python
X_new_poly
```

```python
y_new = regression.predict(X_new_poly)
plt.plot(X_new, y_new, "r-", linewidth = 2, label="Predictions")
plt.plot(X_train, y_train, "b.", label = "Training points")
plt.plot(X_test, y_test, "g.", label ="Testing points")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

# polynomial pipeline
- basically we define a function for linear regression and it takes degree and model gives its prediction for test data and also the required parameters

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_new = np.linspace(-3,3,200).reshape(200,1)
    poly_features = PolynomialFeatures(degree = degree,include_bias = True)
    lin_reg = LinearRegression()
    poly_regression = Pipeline([
        ("poly_features",poly_features),
        ("line_re", lin_reg)
    ])

    poly_regression.fit(X_train, y_train) ## polynomial and fit of linear regression
    y_pred_new = poly_regression.predict(X_new)

    # ploting predictions
    plt.plot(X_new, y_pred_new, 'r', label = "Degree" + str(degree), linewidth = 3)
    plt.plot(X_train, y_train, "b.", linewidth=3)
    plt.plot(X_test, y_test, "g.", linewidth = 3)
    plt.legend(loc = "upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4,0,10])
    plt.show()
```

```python
poly_regression(6)
```
